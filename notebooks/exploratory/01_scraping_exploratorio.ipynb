Código de teste inicial

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.webdriver import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Caminho completo para o seu chromedriver.exe
chromedriver_path = r'C:\Users\vbitu\OneDrive\Documentos\One Drive Backup\OneDrive\Estudos\Projetos e Desafios\Fake News ETL Project\Páginas HTML\Chrome Driver\chromedriver-win64\chromedriver-win64\chromedriver.exe'

# Configura o Chrome
options = Options()
options.add_argument("--start-maximized")
#usar headless quando for pra rodar mesmo


# Inicia o ChromeDriver
service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service, options=options)

# Abre a página de checagens
driver.get("https://www.aosfatos.org/noticias/?formato=checagem")

# Aguarda o carregamento dinâmico
try:
    WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "div.group.relative.size-max"))
    )
except:
    print("Tempo de espera excedido. Conteúdo não carregou.")

# Localizando os cards principais de notícias
cards = driver.find_elements(By.CSS_SELECTOR, "div.group.relative.size-max")

# Loop principal de extração
for card in cards[:12]:  
    try:
        # Extrair o link
        link_element = card.find_element(By.CSS_SELECTOR, "a")
        href = link_element.get_attribute("href")

        # Extrair o título
        title_element = card.find_element(By.CSS_SELECTOR, "p.block")
        titulo = title_element.text

        # Extrair o veredito
        try:
            veredito_element = card.find_element(By.CSS_SELECTOR, "span.rounded-full")
            veredito = veredito_element.text
        except:
            veredito = "indisponível"

        # Extrair as tags temáticas (podem existir múltiplas)
        tags_container = card.find_elements(By.CSS_SELECTOR, "div.flex a")
        tags = [tag.text for tag in tags_container]

        # Exibir o que foi coletado
        print(f"Link: {href}")
        print(f"Título: {titulo}")
        print(f"Veredito: {veredito}")
        print(f"Tags: {tags}")
        print("-" * 50)

    except Exception as e:
        print(f"Erro ao extrair dados do card: {e}")

# Fecha o navegador
driver.quit()

```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.webdriver import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

chromedriver_path = r'C:\Users\vbitu\OneDrive\Documentos\One Drive Backup\OneDrive\Estudos\Projetos e Desafios\Fake News ETL Project\Páginas HTML\Chrome Driver\chromedriver-win64\chromedriver-win64\chromedriver.exe'

options = Options()
options.add_argument("--start-maximized")
service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service, options=options)

driver.get("https://www.aosfatos.org/noticias/?formato=checagem")

try:
    WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "div.group.relative.size-max"))
    )
except:
    print("Tempo de espera excedido. Conteúdo não carregou.")

cards = driver.find_elements(By.CSS_SELECTOR, "div.group.relative.size-max")

# Loop de diagnóstico
for idx, card in enumerate(cards):
    try:
        veredito_element = card.find_element(By.CSS_SELECTOR, "span.rounded-full")
        veredito = veredito_element.text
        print(f"Card {idx+1}: Veredito encontrado -> {veredito}")
    except:
        print(f"Card {idx+1}: Sem veredito encontrado")

driver.quit()
```
