```python
# ----------------------
# ETAPA 1 - INSTALAR AS BIBLIOTECAS NECESSÁRIAS NO COLAB
# ----------------------

!pip install bertopic[visualization]
!pip install umap-learn
!pip install hdbscan
!pip install sentence-transformers

# ----------------------
# ETAPA 2 - IMPORTAÇÃO DAS BIBLIOTECAS
# ----------------------

import pandas as pd
from bertopic import BERTopic
import matplotlib.pyplot as plt
import plotly.io as pio
pio.renderers.default = "colab"

# ----------------------
# ETAPA 3 - CARREGAR O DATASET
# ----------------------

# Antes de rodar, faça upload do arquivo dados_unificados_aosfatos.csv via aba Files do Colab
caminho_csv = "/content/dados_unificados_aosfatos.csv"
df = pd.read_csv(caminho_csv)

# Verifica as primeiras linhas
df.head()

# ----------------------
# ETAPA 4 - PREPARAR OS TÍTULOS PARA O BERTopic
# ----------------------

# Remove registros vazios no campo titulo
df = df.dropna(subset=['titulo'])
titulos = df['titulo'].tolist()

# ----------------------
# ETAPA 5 - EXECUTAR O BERTopic USANDO OS TÍTULOS
# ----------------------

model = BERTopic(language="multilingual")
topics, probs = model.fit_transform(titulos)

# ----------------------
# ETAPA 6 - EXIBIR OS TÓPICOS GERADOS
# ----------------------

topicos_gerados = model.get_topic_info()
topicos_gerados.head(10)

# ----------------------
# ETAPA 7 - VISUALIZAR OS TÓPICOS
# ----------------------

model.visualize_barchart(top_n_topics=10).show()
model.visualize_topics().show()

# ----------------------
# ETAPA 8 - EXPORTAR OS RESULTADOS
# ----------------------

df['topico_bertopic_titulo'] = topics

caminho_saida = "/content/dados_unificados_com_topicos_titulos.csv"
df.to_csv(caminho_saida, index=False, encoding='utf-8-sig')

print("Arquivo exportado com sucesso: dados_unificados_com_topicos_titulos.csv")

# ----------------------
# FIM DO PIPELINE DE EXPLORAÇÃO TEMÁTICA POR TÍTULOS
# ----------------------
```
