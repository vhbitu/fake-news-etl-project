```python
import pandas as pd

# 1. Carregar os dois datasets
df_listagem = pd.read_csv("C:/Users/vbitu/projects/fake-news-etl-project/data/raw/dados_listagem_aosfatos.csv")
df_detalhes = pd.read_csv("C:/Users/vbitu/projects/fake-news-etl-project/data/processed/dados_detalhados_aosfatos.csv")

# 2. Verificar visualmente (opcional)
print("Listagem:")
print(df_listagem.head())
print("Detalhes:")
print(df_detalhes.head())

# 3. Realizar o merge baseado no campo 'link'
df_unificado = pd.merge(df_listagem, df_detalhes, on="link", how="inner")

# 4. Visualizar o resultado final
print("Dataset unificado:")
print(df_unificado.head())
print(f"Formato final: {df_unificado.shape}")

# 5. Salvar o dataset unificado
df_unificado.to_csv("C:/Users/vbitu/projects/fake-news-etl-project/data/processed/dados_unificados_aosfatos.csv", index=False, encoding="utf-8-sig")

print("Merge e salvamento concluídos com sucesso!")
```
