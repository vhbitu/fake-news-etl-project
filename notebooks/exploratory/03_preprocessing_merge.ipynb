```python
import pandas as pd
from pathlib import Path

# Caminho base do projeto (ajuste conforme a localização do notebook)
BASE_DIR = Path().resolve().parent.parent

# Caminhos dos arquivos
LISTAGEM_PATH = BASE_DIR / "data" / "raw" / "dados_listagem_aosfatos.csv"
DETALHES_PATH = BASE_DIR / "data" / "raw" / "dados_detalhados_aosfatos.csv"  # <-- Corrigido!
OUTPUT_PATH = BASE_DIR / "data" / "processed" / "dados_unificados_aosfatos.csv"

# 1. Carregar os datasets
df_listagem = pd.read_csv(LISTAGEM_PATH)
df_detalhes = pd.read_csv(DETALHES_PATH)

# 2. Merge pelo campo 'link'
df_unificado = pd.merge(df_listagem, df_detalhes, on="link", how="inner")

# 3. Remover duplicatas, se houver
df_unificado.drop_duplicates(subset="link", inplace=True)

# 4. Salvar dataset unificado
df_unificado.to_csv(OUTPUT_PATH, index=False, encoding="utf-8-sig")

# 5. Verificação rápida
print("Merge e salvamento concluídos com sucesso!")
print(f"Formato final: {df_unificado.shape}")
df_unificado.head()

```
