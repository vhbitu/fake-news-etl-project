# 04 – Análise Exploratória Inicial (EDA)

Este notebook realiza uma análise exploratória dos dados unificados coletados do site [Aos Fatos](https://www.aosfatos.org).  
As etapas incluem:

- Conversão de datas para formato datetime
- Verificação de campos nulos
- Visualização da distribuição temporal das checagens publicadas

Essa etapa é essencial para entender o comportamento dos dados antes de aplicar modelos de NLP ou classificações.


```python
from pathlib import Path
import pandas as pd
import dateparser

BASE_DIR = Path().resolve().parent.parent
INPUT_PATH = BASE_DIR / "data" / "processed" / "dados_unificados_aosfatos.csv"

df_unificado = pd.read_csv(INPUT_PATH)
df_unificado['data_publicacao'] = df_unificado['data_publicacao'].apply(lambda x: dateparser.parse(x, languages=['pt']))
```

```python
import matplotlib.pyplot as plt

# Agrupamento por mês usando a coluna 'data_publicacao'
df_grouped = df_unificado.set_index('data_publicacao').resample('M').size()

# Plot da distribuição temporal
plt.figure(figsize=(15, 6))
df_grouped.plot(marker='o', linestyle='-')
plt.title("Distribuição temporal das checagens")
plt.xlabel("Data de Publicação")
plt.ylabel("Número de checagens no mês")
plt.grid(True)
plt.tight_layout()
plt.show()
```
