```python
import pandas as pd
import dateparser

# Carregar o dataset unificado
df_unificado = pd.read_csv(r'C:\Users\vbitu\projects\fake-news-etl-project\data\processed\dados_unificados_aosfatos.csv')

# Conversão de data_publicacao usando dateparser (que entende português)
df_unificado['data_publicacao'] = df_unificado['data_publicacao'].apply(lambda x: dateparser.parse(x, languages=['pt']))

# Verificação rápida
print(df_unificado.head())
print(df_unificado.info())
print(df_unificado.isnull().sum())
```

```python
import matplotlib.pyplot as plt

# Agrupando por mês para reduzir ruído 
df_grouped = df_unificado.groupby(pd.Grouper(key='data_publicacao', freq='M')).size()

# Plotando a distribuição temporal
plt.figure(figsize=(15,6))
df_grouped.plot()
plt.title("Distribuição temporal das checagens")
plt.xlabel("Data de Publicação")
plt.ylabel("Número de checagens no mês")
plt.grid()
plt.show()
```
