```python
# ===============================
# Engenharia Temática - NLP (Bloco 3)
# ===============================

import pandas as pd
import re

# Carregando o dataset com o texto limpo
df = pd.read_csv(r'C:\Users\vbitu\projects\fake-news-etl-project\data\processed\dados_unificados_aosfatos_limpo.csv')

# ===============================
# Definindo palavras-chave temáticas
# ===============================

temas = {
    'saude_publica': ['covid', 'vacina', 'virus', 'doenca', 'hospital', 'pandemia', 'infectado', 'epidemia', 'tratamento', 'imunidade'],
    'meio_ambiente': ['clima', 'aquecimento', 'desmatamento', 'floresta', 'poluicao', 'amazonia', 'carbono', 'emissao', 'desastre', 'seca'],
    'politica': ['eleicao', 'governo', 'congresso', 'presidente', 'senado', 'ministerio', 'politico', 'partido', 'campanha'],
    'fake_classico': ['boato', 'corrente', 'mentira', 'falso', 'desmentido', 'fake', 'fraude', 'enganoso', 'factcheck'],
}

# ===============================
# Função de detecção temática
# ===============================

def detectar_tema(texto, palavras_chave):
    texto = texto.lower()
    return int(any(palavra in texto for palavra in palavras_chave))

# Criando as novas colunas temáticas
for tema, palavras in temas.items():
    df[tema] = df['texto_limpo'].apply(lambda x: detectar_tema(x, palavras))

# ===============================
# Visualização básica dos temas
# ===============================

print(df[list(temas.keys())].sum())

# ===============================
# Salvamento do dataset temático
# ===============================

df.to_csv(r'C:\Users\vbitu\projects\fake-news-etl-project\data\processed\dados_aosfatos_tematico.csv', index=False)
```
